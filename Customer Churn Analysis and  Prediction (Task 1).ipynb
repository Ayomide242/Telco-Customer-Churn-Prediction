````python
#Remove duplicates
df = df.drop_duplicates()

# Fill missing values with the mean 
df.fillna(df['TotalCharges'].mean(), inplace = True)
````

````python
# Data Shape after removing duplicates
print(df.shape)
# Confirm if missing values have been replaced
df.isna().sum()
````

```The initial dataset comprised 7,043 observations and 21 variables, including 16 categorical and 4 numerical variables. During the data cleaning process, 22 duplicate records were identified and subsequently removed. Additionally, 11 missing values were detected in the TotalCharges column and were imputed with the mean value of that column. Following these preprocessing steps and the designation of customerID as the dataset index, the cleaned dataset consisted of 7,021 observations and 20 variables.```
